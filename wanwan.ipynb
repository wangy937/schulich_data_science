```python
initial_energy = 1220
max_energy = 1820
battle_cost = 800
total_energy_received = 0
total_energy = initial_energy
received_300_once = False
log = []
step = 0

def can_receive(amount):
    return total_energy < 1000 and total_energy + amount <= max_energy

while True:
    step += 1

    # Try to battle if energy is enough
    if total_energy >= battle_cost:
        total_energy -= battle_cost
        log.append(f"Step {step}: Battle (-{battle_cost}) → Energy = {total_energy}")
        continue

    # Decide which energy to receive
    next_receive = None
    if not received_300_once and can_receive(300):
        next_receive = 300
    elif received_300_once:
        if can_receive(820) and total_energy_received + 820 <= 5600:
            next_receive = 820
        elif can_receive(600) and total_energy_received + 600 <= 5600:
            next_receive = 600

    # Final allowed receive if we're near the cap
    if not next_receive:
        if not received_300_once and total_energy < 1000 and total_energy + 300 <= max_energy:
            next_receive = 300
        elif received_300_once:
            if total_energy < 1000 and total_energy + 820 <= max_energy and total_energy_received < 5600:
                next_receive = 820
            elif total_energy < 1000 and total_energy + 600 <= max_energy and total_energy_received < 5600:
                next_receive = 600

            # Allow final receive even if it causes going over 5600
            if total_energy < 1000:
                if total_energy + 820 <= max_energy and total_energy_received < 5600:
                    next_receive = 820
                elif total_energy + 600 <= max_energy and total_energy_received < 5600:
                    next_receive = 600

    # If no valid energy to receive and can't battle → done
    if not next_receive:
        break

    # Apply receive
    total_energy += next_receive
    total_energy_received += next_receive
    if next_receive == 300:
        received_300_once = True
    log.append(f"Step {step}: Receive {next_receive} (+{next_receive}) → Energy = {total_energy}, Total received = {total_energy_received}")

    # Stop if final receive pushed us over 5600
    if total_energy_received > 5600:
        break

# Final stats
print("\n".join(log))
print("\nFinal Summary:")
print(f"Total energy received: {total_energy_received}")
print(f"Total battles possible: {total_energy // battle_cost}")
print(f"Remaining energy: {total_energy}")
```

```python
initial_energy = 1220
max_energy = 1820
total_energy_collected = 0
max_total_collectable = 5600
current_energy = initial_energy
collected_300 = False
collected_600_enabled = False
actions = []
battle_count = 0

def can_collect(amount):
    return current_energy + amount <= max_energy and current_energy < 1000

def do_battle():
    global current_energy, battle_count
    if current_energy >= 800:
        current_energy -= 800
        battle_count += 1
        actions.append(f"战斗 -800 → 当前能量: {current_energy}")

while True:
    # 战斗优先，只要够800就去战斗
    while current_energy >= 800:
        do_battle()

    # 如果领取后将超出5600并不是最后一次，则停止
    if total_energy_collected >= max_total_collectable:
        break

    # 优先领取300点（只能一次）
    if not collected_300 and can_collect(300):
        current_energy += 300
        total_energy_collected += 300
        collected_300 = True
        collected_600_enabled = True
        actions.append(f"领取300 → 当前能量: {current_energy}，累计领取: {total_energy_collected}")
        continue

    # 尝试领取820点
    if can_collect(820) and total_energy_collected + 820 <= max_total_collectable:
        current_energy += 820
        total_energy_collected += 820
        actions.append(f"领取820 → 当前能量: {current_energy}，累计领取: {total_energy_collected}")
        continue

    # 尝试领取600点（前提：已领过300）
    if collected_600_enabled and can_collect(600) and total_energy_collected + 600 <= max_total_collectable:
        current_energy += 600
        total_energy_collected += 600
        actions.append(f"领取600 → 当前能量: {current_energy}，累计领取: {total_energy_collected}")
        continue

    # 如果领取820会超出，但这是最后一次，我们允许
    if can_collect(820) and total_energy_collected < max_total_collectable:
        current_energy += 820
        total_energy_collected += 820
        actions.append(f"最后一次领取820（超出） → 当前能量: {current_energy}，累计领取: {total_energy_collected}")
        break

    # 如果领取600会超出，但这是最后一次，我们允许
    if collected_600_enabled and can_collect(600) and total_energy_collected < max_total_collectable:
        current_energy += 600
        total_energy_collected += 600
        actions.append(f"最后一次领取600（超出） → 当前能量: {current_energy}，累计领取: {total_energy_collected}")
        break

    # 如果无法再领取，退出循环
    break

# 战斗结算
while current_energy >= 800:
    do_battle()

# 输出所有动作和总结
print("\n操作记录：")
for act in actions:
    print(act)

print("\n总结：")
print(f"累计领取能量: {total_energy_collected}")
print(f"最终剩余能量: {current_energy}")
print(f"最多对战次数: {battle_count}")
```

```python
import gurobipy as gp
from gurobipy import GRB

T = 30  # 时间步长
initial_energy = 1220
max_energy = 1820
max_total_collect = 5600

model = gp.Model("energy_strategy")

# Variables
x_300 = model.addVars(T, vtype=GRB.BINARY, name="x_300")
x_600 = model.addVars(T, vtype=GRB.BINARY, name="x_600")
x_820 = model.addVars(T, vtype=GRB.BINARY, name="x_820")
b = model.addVars(T, vtype=GRB.BINARY, name="battle")
e = model.addVars(T + 1, vtype=GRB.CONTINUOUS, name="energy")
z = model.addVars(T, vtype=GRB.BINARY, name="z_below_1000")  # 1 if energy[t] < 1000

# Initial energy
model.addConstr(e[0] == initial_energy)

# At most one type of energy can be collected per timestep
for t in range(T):
    model.addConstr(x_300[t] + x_600[t] + x_820[t] <= 1)

# Energy dynamics
for t in range(T):
    gain = 300 * x_300[t] + 600 * x_600[t] + 820 * x_820[t]
    loss = 800 * b[t]
    model.addConstr(e[t + 1] == e[t] - loss + gain)
    model.addConstr(e[t + 1] <= max_energy)

# Big-M constraint to prevent energy collection if energy[t] >= 1000
M = 10000
for t in range(T):
    model.addConstr(e[t] <= 999 + M * (1 - z[t]))
    model.addConstr(x_300[t] + x_600[t] + x_820[t] <= z[t])  # 只能在 z[t] == 1 的时候领取能量

# 300 only once
model.addConstr(gp.quicksum(x_300[t] for t in range(T)) <= 1)

# 600 only allowed after 300 was used
for t in range(T):
    model.addConstr(x_600[t] <= gp.quicksum(x_300[k] for k in range(t)))

# Total collected energy <= 5600 (but allow one overstep for last pickup)
collect_total = gp.quicksum(
    300 * x_300[t] + 600 * x_600[t] + 820 * x_820[t] for t in range(T)
)
model.addConstr(collect_total <= max_total_collect + 820)

# Objective: maximize battles
model.setObjective(gp.quicksum(b[t] for t in range(T)), GRB.MAXIMIZE)

# Optimize
model.optimize()

# Output results
print("\n操作记录：")
for t in range(T):
    if x_300[t].x > 0.5:
        print(f"Step {t}: 领取300")
    if x_600[t].x > 0.5:
        print(f"Step {t}: 领取600")
    if x_820[t].x > 0.5:
        print(f"Step {t}: 领取820")
    if b[t].x > 0.5:
        print(f"Step {t}: 战斗一次")
    print(f"    当前能量: {e[t].x:.1f}")

print(f"\n总对战次数：{sum(b[t].x for t in range(T))}")
print(f"累计领取能量：{collect_total.getValue():.0f}")
```
