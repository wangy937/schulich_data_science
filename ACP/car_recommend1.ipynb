```python
# 1. Create a new column 'is_moving' to indicate whether the vehicle is moving
# If 'speed' is -1, the vehicle is considered stopped (0), otherwise it is moving (1)
car_df['is_moving'] = car_df['speed'].apply(lambda x: 0 if x == -1 else 1)

# 2. Estimate the maximum battery capacity for each vehicle
# The highest recorded 'battery_range' for each 'vehicle_id' is used as an estimate of its battery capacity
car_df['estimated_battery_capacity'] = car_df.groupby('vehicle_id')['battery_range'].transform('max')

# 3. Create a low battery warning feature
# If 'battery_range' is less than 50, flag as 1 (low battery), otherwise 0
car_df['first_low_battery_warning'] = car_df['battery_range'].apply(lambda x: 1 if x < 50 else 0)

# 4. Create a low battery warning feature
# If 'battery_range' is less than 50, flag as 1 (low battery), otherwise 0
car_df['second_low_battery_warning'] = car_df['battery_range'].apply(lambda x: 1 if x < 40 else 0)

# 5. Create a low battery warning feature
# If 'battery_range' is less than 50, flag as 1 (low battery), otherwise 0
car_df['third_low_battery_warning'] = car_df['battery_range'].apply(lambda x: 1 if x < 30 else 0)



```

```python

# Create a binary variable 'battery_heat_warning' to indicate when to turn on battery heating
# If 'outside_temp' is less than 5°C, set it to 1 (warning to turn on battery heat), otherwise 0
car_df['battery_heat_warning'] = car_df['outside_temp'].apply(lambda x: 1 if x < 5 else 0)
```

```python
# Create a binary variable 'battery_check_warning' to remind the owner to check the battery
# If 'max_range_charge_counter' reaches 20, set to 1 (first warning), otherwise 0
car_df['battery_check_warning_20'] = car_df['max_range_charge_counter'].apply(lambda x: 1 if x == 20 else 0)

# Create a binary variable 'battery_check_warning' for a second reminder at 30 charges
# If 'max_range_charge_counter' reaches 30, set to 1 (second warning), otherwise 0
car_df['battery_check_warning_30'] = car_df['max_range_charge_counter'].apply(lambda x: 1 if x == 30 else 0)

# Create a binary variable 'battery_check_warning' for a second reminder at 40 charges
# If 'max_range_charge_counter' reaches 40, set to 1 (second warning), otherwise 0
car_df['battery_check_warning_40'] = car_df['max_range_charge_counter'].apply(lambda x: 1 if x == 40 else 0)
```
